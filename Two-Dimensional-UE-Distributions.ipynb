```python
import numpy as np
import matplotlib.pyplot as plt
# increase plot size
plt.rcParams['figure.figsize'] = [10, 10]
# increase resolution
plt.rcParams['figure.dpi'] = 100
# set font size
plt.rcParams.update({'font.size': 18})
# set font as Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
```

```python

UE_num = 1000
Area_len = np.pi
Sigma = Area_len/7
x = np.random.normal(0, Sigma, UE_num)
y = np.random.normal(0, Sigma, UE_num)


# set x label as \phi angle, y label as \theta angle, use latex, from -\pi/2 to \pi/2, unit is rad


# plot the UE positions
plt.scatter(x, y)


plt.xlabel(r'$\phi$ (rad)')
plt.ylabel(r'$\theta$ (rad)')
# plt.xlabel('x (m)')
# plt.ylabel('y (m)')
# set xlim and ylim
plt.xlim(-Area_len/2, Area_len/2)
plt.ylim(-Area_len/2, Area_len/2)


# set legend with text "UE distribution"
plt.legend(['UE distribution'])

plt.show()
```
